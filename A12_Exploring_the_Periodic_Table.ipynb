```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# **Task 1.** Create a scatterplot to compare the Period in which an element occurs (“Period”) to its Number of Electron Shells (NmberofShells”). Calculate the Pearson correlation coefficient for these two variables. Explain the relationship between Period and Number of Electron Shells. 

```python
new = ptb.copy()
new = new[['Period','NumberofShells']]
new.dropna(inplace=True)

fig, ax = plt.subplots(figsize=(20,5))
b = sns.scatterplot(
    data=new, x='Period', y="NumberofShells", 
    s=300, palette="deep");

b.axes.set_title("Scatter plot relationship btw Period and NumberofShells",fontsize=24)
b.set_xlabel(".Period ",fontsize=20)
b.set_ylabel(".NumberofShells",fontsize=20)
b.tick_params(labelsize=15)
plt.grid()
plt.show()
print("Pearson Corelation of Period and NumberofShells is ",ptb['Period'].corr(ptb['NumberofShells'],method='pearson'))
print("Its means the corelation is positive Between Period and NumberofShell if Period go increase then numberofshell will also increase ")
```

# **Task 2.** Create a scatterplot to compare how the Number of Valence Electrons (“NumberofValence”) varies by group (“Group”). Explain the observed pattern.

```python
gp= ptb.copy()
gp = gp[['NumberofValence','Group']]
gp.dropna(inplace=True)

fig, ax = plt.subplots(figsize=(12,8))
b = sns.scatterplot(
    data=gp, x='NumberofValence', y="Group", 
    s=100, );

b.axes.set_title("Scatter plot relationship btw NumberofValence Electrons  and group by",fontsize=18)
b.set_xlabel("NumberofValence ",fontsize=20)
b.set_ylabel("Group",fontsize=20)
b.tick_params(labelsize=15)
ax.set_ylim(0,19)
plt.grid()
plt.yticks(np.arange(0,19,1))
plt.show();
```

# **Task 5.** Create a scatterplot to compare the Atomic Number (“AtomicNumber”) and Atomic Radius (“AtomicRadius”) for just elements in Period 5 (“Period”). Explain the observed relationship. 

```python
at = ptb.copy()
at = at[['Period','AtomicNumber','AtomicRadius']]
at = at[at['Period'] == 5]
at.dropna(inplace=True)
fig, ax = plt.subplots(figsize=(20,7))
b = sns.scatterplot(
    data=at, x='AtomicNumber', y="AtomicRadius",hue= 'Period',palette='deep', s=100 );

b.axes.set_title("Scatter plot relationship btw AtomicNumber and AtomicMass in period 5",fontsize=18)
b.set_xlabel("AtomicNumber ",fontsize=20)
b.set_ylabel("AtomicRadius",fontsize=20)
b.tick_params(labelsize=15)
plt.grid()
plt.show();
```
