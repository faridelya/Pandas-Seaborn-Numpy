# **Q15**
-  Does Marvel of DC have a higher percentage of movies with a PG-13
rating vs. all other ratings (“mpa_rating”)?

```python
# in PG-13
pg= df[df['mpa_rating'] == 'PG-13']
dc = pg[pg['entity']=='DC'].shape[0]
mar =pg[pg['entity']=='MARVEL'].shape[0]

dc_per = 100*dc/len(pg)
mar_per = 100*mar/len(pg)

print("in PG-13 the percentage of DC is {} and MARVEL is {}".format(dc_per,mar_per))
```

```python
# in other
other = df[df['mpa_rating'] != 'PG-13']
dc = other[other['entity']=='DC'].shape[0]
mar =other[other['entity']=='MARVEL'].shape[0]

dc_per = dc/len(pg)*100
mar_per = mar/len(pg)*100
print("in PG-13 the percentage of DC is {} and MARVEL is {}".format(dc_per,mar_per))
```
