**Task 6: Create a line graph showing the estimated mass of the least massive planet
detected each year. Discuss the trend observed. Note that there are some missing
values.**

```python
import numpy as np
year =[]
least_mass= []
for i in Exp['Discovery Year'].unique():
    # print("year ", i)
    fil = Exp[Exp['Discovery Year'] == i]
    year.append(i)
    lst_ms = fil['Mass'].min()
    # print("lst mass ",lst_ms)
    least_mass.append(lst_ms)

df = pd.DataFrame({"Year":year, "least_massive":least_mass})
df.sort_values(by='Year', ascending=True, inplace=True)
df.reset_index(drop=True,inplace=True)
fig , ax = plt.subplots(figsize=(23,7))
plt.plot(df['Year'],df['least_massive'],marker='o',markersize=8, linewidth=2, linestyle = 'dashed',color='green',axes=ax)
ax.set_title("Discovery of Least massive planet each year",fontsize=25)
ax.set_ylabel("Mass",fontsize=20)
ax.set_xlabel("Discovery Year",fontsize=20)
ax.set_xlim(1987,2022)
plt.xticks(np.arange(1989,2022,1))
plt.xticks(rotation = 45) 
plt.grid()
ax.tick_params(labelsize=15)
for index in range(len(df['Year'])):
  ax.text(df['Year'][index],df['least_massive'][index], df['least_massive'][index], size=12, rotation_mode='anchor',rotation='vertical')
plt.show()
```

**Task 7: Create a line graph showing the estimated mass of the least massive planet
detected each year with separate lines for each of the four detection methods. Discuss
the trend observed. Note that there are some missing values.**

```python

# selected four method
four = Exp[Exp['Discovery Method'].isin(['Radial Velocity', 'Imaging', 'Eclipse Timing Variations','Transit'])]

lst = []
mthd = []



for m in four['Discovery Method'].unique():
    fl = four[four['Discovery Method'] == m]
    for y in fl['Discovery Year'].unique():
        f2 = fl[fl['Discovery Year']== y]
        lst_ms = f2['Mass'].min()
        lst.append([y,lst_ms])
        mthd.append(m)

     
df = pd.DataFrame(lst , columns=["year","least_mass"])
a =pd.DataFrame({"method":mthd})
df = pd.concat([df,a], axis=1)
df.sort_values(by='year', ascending=True, inplace=True)
df.reset_index(drop=True,inplace=True)

fig , ax = plt.subplots(figsize=(18,8))
sns.lineplot(data=df, x="year", y="least_mass", hue='method',palette = 'deep',ax=ax,  markers=True,dashes=False)
ax.set_title("Discovery of Least massive planet each year in four method",fontsize=25)
ax.set_ylabel("Mass",fontsize=20)
ax.set_xlabel("Discovery Year",fontsize=20)
# ax.set_xlim(1987,2022)
# plt.xticks(np.arange(1989,2022,1))
plt.grid()
ax.tick_params(labelsize=15)
plt.show();
```

```python
first = df[df['method'] == 'Radial Velocity']
first.sort_values(by='year', ascending=True, inplace=True)
first.reset_index(drop=True,inplace=True)
second = df[df['method']== 'Imaging']
second.sort_values(by='year', ascending=True, inplace=True)
second.reset_index(drop=True,inplace=True)
thrd = df[df['method']== 'Eclipse Timing Variations']
thrd.sort_values(by='year', ascending=True, inplace=True)
thrd.reset_index(drop=True,inplace=True)
frth =df[df['method']== 'Transit']
frth.sort_values(by='year', ascending=True, inplace=True)
frth.reset_index(drop=True,inplace=True)

fig , ax = plt.subplots(figsize=(18,8))
plt.plot(first['year'],first['least_mass'],marker='o',markersize=8, linewidth=2, linestyle = 'dashed',axes=ax, )
plt.plot(second['year'],second['least_mass'],marker='o',markersize=8, linewidth=2, linestyle = 'dashed',axes=ax, )
plt.plot(thrd['year'],thrd['least_mass'],marker='o',markersize=8, linewidth=2, linestyle = 'dashed',axes=ax, )
plt.plot(frth['year'],frth['least_mass'],marker='o',markersize=8, linewidth=2, linestyle = 'dashed',axes=ax, );
plt.grid()
ax.set_ylabel("Mass",fontsize=20)
ax.set_xlabel("Discovery Year",fontsize=20)
plt.show();
```
